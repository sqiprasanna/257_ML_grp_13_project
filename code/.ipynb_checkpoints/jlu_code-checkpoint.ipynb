```python
def cross_val(X_train, Y_train, X_test, Y_test, model):
    num_cv = 5

    scoring = ['accuracy', 'precision', 'recall']
    cv = cross_validate(model, X_train, Y_train, cv=num_cv, return_train_score=True, scoring=scoring, n_jobs=-1)
    
    train_scores = []
    test_scores = []
    
    for metric in scoring:
        train_scores.append(np.mean(cv[f"train_{metric}"]))
        test_scores.append(np.mean(cv[f"test_{metric}"]))
        
    mean_scores = pd.DataFrame((train_scores, test_scores), columns=scoring, index=["train", "test"])
    
    return cv, mean_scores
```

```python
def plot_cross_val(cv):    
    metrics = ["accuracy", "precision", "recall"]
    colors = ["r", "g", "b"]
    for time_ in ["train", "test"]:
        for num, metric in enumerate(metrics):
            indexer = f"{time_}_{metric}"
            plt.plot(np.arange(5), cv[indexer], color=colors[num])
        plt.legend(metrics, loc="upper left")
        plt.title(f"Scores during {time_} time")
        plt.show()
```

```python
def pred(X_train, Y_train, X_test, Y_test, model):
    clf = model.fit(X_train, Y_train)
    pred = model.predict(X_train)
    pred2 = model.predict(X_test)
    
    print("Training accuracy: ", accuracy_score(Y_train, pred))
    print("Testing accuracy: ", accuracy_score(Y_test, pred2))
    print()
    
    print("Training precision: ", precision_score(Y_train, pred))
    print("Testing precision: ", precision_score(Y_test, pred2))
    print()
    
    print("Training recall: ", recall_score(Y_train, pred))
    print("Testing recall: ", recall_score(Y_test, pred2))
    print()

    conf_mat = confusion_matrix(y_true=Y_train, y_pred=pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat)
    disp.plot()
    
    conf_mat2 = confusion_matrix(y_true=Y_test, y_pred=pred2)
    disp2 = ConfusionMatrixDisplay(confusion_matrix=conf_mat2)
    disp2.plot()
```
